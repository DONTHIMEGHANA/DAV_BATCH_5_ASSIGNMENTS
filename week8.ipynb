```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# -----------------------------
# 1️⃣ Create Synthetic Dataset
# -----------------------------
n = 50  # number of months

data = {
    'Sales': np.random.normal(50000, 8000, n),
    'Ad_Spend': np.random.normal(10000, 2000, n),
    'Discounts': np.random.normal(8, 2, n),
    'Customer_Visits': np.random.normal(2000, 300, n),
    'Online_Reviews': np.random.normal(4.0, 0.5, n)
}

df = pd.DataFrame(data)

# Add relationships: Sales increases with Ad_Spend and Discounts slightly
df['Sales'] = (df['Ad_Spend'] * 4.5) + (df['Discounts'] * 1000) + np.random.normal(0, 5000, n)

# Display first few rows
print("Preview of dataset:")
print(df.head(), "\n")

# -----------------------------
# 2️⃣ Compute Correlations
# -----------------------------
corr_matrix = df.corr()
print("Correlation Matrix:\n")
print(corr_matrix, "\n")

# -----------------------------
# 3️⃣ Plot Heatmap
# -----------------------------
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap: Sales & Marketing Variables')
plt.show()

# -----------------------------
# 4️⃣ Check correlation with Sales
# -----------------------------
sales_ad_corr = corr_matrix.loc['Sales', 'Ad_Spend']
sales_disc_corr = corr_matrix.loc['Sales', 'Discounts']

print(f"Correlation between Sales and Ad_Spend: {sales_ad_corr:.3f}")
print(f"Correlation between Sales and Discounts: {sales_disc_corr:.3f}")

# Determine stronger relationship
if abs(sales_ad_corr) > abs(sales_disc_corr):
    print("\n✅ Advertising Spend has a stronger correlation with Sales.")
else:
    print("\n✅ Discounts have a stronger correlation with Sales.")
```

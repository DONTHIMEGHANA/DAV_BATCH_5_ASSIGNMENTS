```python
import plotly.express as px
import pandas as pd

# Step 1: Data
data = {
    'State': ['Maharashtra', 'Kerala', 'Delhi', 'Karnataka', 'Tamil Nadu'],
    'Cases': [320, 250, 150, 180, 200]
}
df = pd.DataFrame(data)

# Step 2: Load India states GeoJSON
geojson_url = 'https://raw.githubusercontent.com/datameet/maps/master/States/Admin2.json'

# Step 3: Create Choropleth
fig = px.choropleth(
    df,
    geojson=geojson_url,
    featureidkey='properties.ST_NM',   # matches GeoJSON field for state name
    locations='State',
    color='Cases',
    color_continuous_scale='Reds',
    title='COVID-19 Cases in Indian States (2020)'
)

fig.update_geos(fitbounds="locations", visible=False)
fig.show()
```
